```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
dataset = pd.read_csv('./Datasets/dataset_cleaned.csv')
df = dataset.copy()
df.head()
```

```python
# Veri setindeki değişken ve gözlem sayısı
print("Gözlem sayısı : " ,len(df))
print("Değişken sayısı : ", len(df.columns))
```

```python
f,ax=plt.subplots(1,2,figsize=(18,8))
df['Label_Num'].value_counts().plot.pie(explode=[0,0.1],autopct='%1.1f%%',ax=ax[0],shadow=True)
ax[0].set_title('Label_Num')
ax[0].set_ylabel('')
sns.countplot('Label_Num',data=df,ax=ax[1])
ax[1].set_title('Label_Num')
plt.show()
```

```python
f,ax=plt.subplots(1,2,figsize=(18,8))
df['Label_Num'].value_counts().plot.pie(explode=[0,0.1],autopct='%1.1f%%',ax=ax[0],shadow=True)
ax[0].set_title('Label_Num')
ax[0].set_ylabel('')
sns.countplot('Cat_Num',data=df,ax=ax[1])
ax[1].set_title('Cat_Num')
plt.show()
```

```python
f,ax=plt.subplots(1,2,figsize=(18,8))
df['Label_Num'].value_counts().plot.pie(explode=[0,0.1],autopct='%1.1f%%',ax=ax[0],shadow=True)
ax[0].set_title('Label_Num')
ax[0].set_ylabel('')
sns.countplot('Sub_Cat_Num',data=df,ax=ax[1])
ax[1].set_title('Sub_Cat_Num')
plt.show()
```
